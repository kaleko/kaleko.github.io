```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#array of random x values, and y values are sin of x values w/ some spread
x = np.random.randn(1e5)
y = np.sin(x)+np.random.normal(0,0.1,len(x))
```

```python
#put x- and y- arrays into a dictionary... this seems to be the easiest
#way to turn the two arrays into a pandas dataframe
d = { 'x':x, 'y':y }
df = pd.DataFrame(d)
```

```python
#just to show what the dataframe looks like...
pd.options.display.max_rows = 5
df
```

```python
#the pandas "plot" command. Give it two columns of a dataframe to compare
#some options shown. the cells are hexagons, not squares, interestingly...
#'gridsize' is the number of bins along one axis
df.plot(x='x',y='y',
        kind='hexbin',
        figsize=[10,6],
        title='My Correlation Plot',
        colormap='jet',
        gridsize=100)
```
